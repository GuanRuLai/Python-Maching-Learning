# Data preprocessing

## Import dataset

```python
import os
if not os.path.isdir("HappyML"):
  os.system("git clone https://github.com/cnchi/HappyML.git")

import pandas as pd
df = pd.read_csv("Data.csv")
```

## Split independent variables and dependent variable

```python
import HappyML.preprocessor as pp

X, Y = pp.decomposition(dataset=df, x_columns=[i for i in range(10)], y_columns=[-1])
X = X.values
Y = Y.values
print(X)
print(Y)
```

## Handle missing data

```python
print(df.isnull().sum())
```

## Split training set and testing set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
Y_train[Y_train == 2] = 0
Y_train[Y_train == 4] = 1
Y_test[Y_test == 2] = 0
Y_test[Y_test == 4] = 1
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

# Classification model processing

## Model training

```python
from xgboost import XGBClassifier

classifier = XGBClassifier()
classifier.fit(X_train, Y_train)
```

## Model prediction

```python
Y_pred = classifier.predict(X_test)
Y_pred[Y_pred == 0] = 2
Y_pred[Y_pred == 1] = 4
Y_pred = pd.DataFrame(Y_pred, columns=["pred_Class"])

Y_test[Y_test == 0] = 2
Y_test[Y_test == 1] = 4
Y_test = pd.DataFrame(Y_test, columns=["true_Class"])

df_pred = pd.concat([Y_test, Y_pred], axis=1)
print(df_pred)
```

## Make confusion matrix & Model evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(Y_test, Y_pred)
print(cm)
print(accuracy_score(Y_test, Y_pred))
```
